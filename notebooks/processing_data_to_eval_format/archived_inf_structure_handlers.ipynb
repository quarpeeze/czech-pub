```python
from pathlib import Path
import json
```

```python
# data and file paths
DATA_PATH = Path("data/information-structure/generated")

FILES = [
    "inf-structure-baseline-80-raw.json",
    "inf-structure-explicit-question-80-raw.json",
    "inf-structure-correction-80-raw.json",
]  
```

```python
def assign_ids(rows, cue):
    """
    rows: list of scenario dicts (already loaded)
    cue:  string like "baseline" / "explicit_question" / "correction"
    """
    for i, row in enumerate(rows, start=1):
        cue = cue.replace("_", "-")
        row["scenario_id"] = f"information-structure-{cue}-{i}"
    return rows

# usage:

for f in FILES:

    path = DATA_PATH / f
    path_out = DATA_PATH / f # (DATA_PATH / (f[:-5] + "_with_ids.json"))


    rows = json.load(open(path, "r", encoding="utf-8"))
    cue = rows[0]["cue"]  # or pass explicitly
    rows = assign_ids(rows, cue)
    json.dump(rows, open(path_out, "w", encoding="utf-8"), ensure_ascii=False, indent=2)
```
