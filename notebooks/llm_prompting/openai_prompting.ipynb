```python
from openai import OpenAI
import os
from dotenv import load_dotenv
from pathlib import Path

load_dotenv()
```

```python
API_KEY = os.environ["OPENAI_API_KEY"]
```

```python
client = OpenAI(
  api_key=API_KEY
)
```

```python
test = """
Vyber, co mluvčí svou odpovědí v daném kontextu nepřímo sděluje.

Přečti si kontext, výpověď a možnosti.
Vyber jednu nejlepší odpověď.

Odpověz pouze jedním písmenem: A, B, C nebo D.

Kontext: 
 Y právě přijel z jiného města, aby se viděl s X.

Výpověď:
- A: Baví tě bydlet mimo město?
- B: Život na venkově naprosto miluju.

Co mluvčí B svou odpovědí v daném kontextu nepřímo sděluje?

Možnosti:
A. Ano, baví.
B. Ne, nebaví.
C. Život na cestách bývá někdy dost drahý.
D. Venkov nabízí více prostoru než město.


"""
```

```python


response = client.responses.create(
  model="gpt-5-nano",
  input=test,
  store=True,
  reasoning={"effort": "high"},
  max_output_tokens=10000
)
print(response.output_text)
```
