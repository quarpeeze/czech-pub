```python
import json
from pathlib import Path
```

## old_schema_to_new_schema

```python

def convert_item(old_item: dict) -> dict:
    meta_general = old_item.get("meta_general", {}) or {}
    meta_special = old_item.get("meta_special", {}) or {}

    creation_method = meta_general.get("creation_method")
    if creation_method == "cql_query":
        creation_method = "corpus_mining"

    new_item = {
        "id": old_item["id"],
        "creation_method": creation_method,
        "phenomenon": old_item["phenomenon"],
        "category": old_item["category"],
        "context": old_item["context"],
        "utterance": old_item["utterance"],
        "question": old_item["question"],
        "options": old_item["options"],
        "correct_option": old_item["correct_option"],
    }

    metadata = {
        "corpus": meta_general.get("corpus"),
        "genre": meta_general.get("genre"),
        "trigger": meta_special.get("trigger"),
        "negation": meta_special.get("negation"),
    }
    metadata = {k: v for k, v in metadata.items() if v not in (None, "", [], {})}

    new_item["metadata"] = metadata
    return new_item


def convert_json_file(in_path, out_path):
    in_path = Path(in_path)
    out_path = Path(out_path)

    with open(in_path, "r", encoding="utf-8") as f:
        data = json.load(f)

    if isinstance(data, list):
        converted = [convert_item(item) for item in data]
    else:
        converted = convert_item(data)

    with open(out_path, "w", encoding="utf-8") as f:
        json.dump(converted, f, ensure_ascii=False, indent=2)

```

```python
old = r"data\presupposition\eval\old_schema\presupposition-wh-question-60.json"
new = r"data\presupposition\eval\new_schema\presupposition-wh-question-60.json"
# example
convert_json_file(old, new)
```

## json to jsonl

```python
import json

file_in = ".json"
file_out = ".jsonl"

def json_to_jsonl(file_in, file_out):
    with open(file_in, "r", encoding="utf-8") as f:
        data = json.load(f)   

    with open(file_out, "w", encoding="utf-8") as f:
        for item in data:
            f.write(json.dumps(item, ensure_ascii=False) + "\n")
```
