```python
import os
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv() 
```

```python
API_KEY = os.environ["ANTHROPIC_API_KEY"]
```

```python
test = """
Vyber, co mluvčí svou odpovědí v daném kontextu nepřímo sděluje.

Přečti si kontext, výpověď a možnosti.
Vyber jednu nejlepší odpověď.

Odpověz pouze jedním písmenem: A, B, C nebo D. Pak pridej vysvetleni.

Kontext: 
 Y právě přijel z jiného města, aby se viděl s X.

Výpověď:
- A: Baví tě bydlet mimo město?
- B: Život na venkově naprosto miluju.

Co mluvčí B svou odpovědí v daném kontextu nepřímo sděluje?

Možnosti:
A. Ano, baví.
B. Ne, nebaví.
C. Život na cestách bývá někdy dost drahý.
D. Venkov nabízí více prostoru než město.


"""
```

```python
test2 = """
Vyber, co mluvčí v daném kontextu nejspíš řekl a jak to formuloval. Představ si, že větu říkáš nahlas.
 
Bylo patrné, že programátor s tím skriptem něco provedl, ale nikdo nevěděl, co s ním přesně udělal.
Marek na to najednou prohlásil:

Možnosti:
A. Smazal programátor skript.
B. Skript smazal programátor.
C. Programátor skript smazal.
D. Při čištění serveru se nepotřebné soubory mažou úplně běžně.
"""
```

```python
client = Anthropic(api_key=API_KEY)

response = client.messages.create(
    model="claude-haiku-4-5",
    max_tokens=16000,
    temperature=0.0,
    system="You are a precise benchmark assistant. Output only the final answer, then add explanation.",
    messages=[
        {
            "role": "user",
            "content": test2
        }
    ],
)

print(response.content[0].text)
print("stop_reason:", response.stop_reason)
print("input_tokens:", response.usage.input_tokens)
print("output_tokens:", response.usage.output_tokens)
```
