## test sheet

```python
def make_test_sheet(phenomena: str, n: int = 10) -> str:

    output = ""
    
    if phenomena == "implicature":
        data_path = ROOT / "data" / "implicature" / "eval" / "eval_circa_indirect_answer_binary_550.json"
    elif phenomena == "presupposition":
        data_path = ROOT / "data" / "presupposition" / "eval" / "presupposition-counterfactual-60.json"
    elif phenomena == "information_structure":
        data_path = ROOT / "data" / "information-structure" / "eval" / "inf_structure_eval_720_v1.json"
    elif phenomena == "scalar_implicature":
        data_path = ROOT / "data" / "implicature" / "eval" / "scalars_eval_100.json"
    data = load_dataset(data_path)

    random_nums = random.sample(range(0, len(data)), n)
    for num in random_nums:
        output+=f"\n\n\n{build_prompt(data[num])}\n\n\n"

    return output
        
```

```python
print(make_test_sheet("presuppositions", 5))
```

```python
print(make_test_sheet("implicature", 10))
```
