- load the merged `eval` and `dev` datasets.
- print simple counts by phenomenon and category.
- inspect example items from different parts of the data.
- render the examples in the same format used by `prompt_builder.py`.
- change the `phenomenon`, `category`, or `index` in the last cells to inspect other items.

```python
# import the small helpers we need
from collections import Counter
from pathlib import Path

from utils.helpers import load_dataset
from utils.prompt_builder import print_question_for_category

ROOT = Path.cwd()
```

```python
def print_dataset_stats(name, items):
    print(f"\n{name}")
    print(f"total items: {len(items)}")

    phenomenon_counts = Counter(item["phenomenon"] for item in items)
    print("\nitems by phenomenon:")
    for phenomenon in sorted(phenomenon_counts):
        print(f"- {phenomenon}: {phenomenon_counts[phenomenon]}")

    print("\nitems by category:")
    category_counts = Counter((item["phenomenon"], item["category"]) for item in items)
    for phenomenon, category in sorted(category_counts):
        count = category_counts[(phenomenon, category)]
        print(f"- {phenomenon} / {category}: {count}")
```

```python
# load the merged eval dataset
eval_path = ROOT / "data" / "eval" / "eval_czech_pub.json"
eval_items = load_dataset(eval_path)

print(f"Loaded {len(eval_items)} items from {eval_path}")
print(f"First item id: {eval_items[0]['id']}")
```

```python
# print simple eval statistics by phenomenon and category
print_dataset_stats("eval dataset", eval_items)
```

```python
# load the merged dev dataset
dev_path = ROOT / "data" / "dev" / "dev_czech_pub.json"
dev_items = load_dataset(dev_path)

print(f"Loaded {len(dev_items)} items from {dev_path}")
print(f"First item id: {dev_items[0]['id']}")
```

```python
# print simple dev statistics by phenomenon and category
print_dataset_stats("dev dataset", dev_items)
```

```python
# print one presupposition example from eval
print_question_for_category(eval_items, phenomenon="presupposition", category="temporal")
```
